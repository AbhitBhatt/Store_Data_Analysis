## DATA CLEANING & PREPARATION¶

```python
df['discounted_price']=df['discounted_price'].str.replace('₹','')
df['discounted_price']=df['discounted_price'].str.replace(',','').astype('float32')
```

```python
df['actual_price']=df['actual_price'].str.replace('₹','').str.replace(',','').astype('float32')
```

```python
df['discount_percentage']=df['discount_percentage'].str.replace('%','').astype('float32')/100.0
```

```python
df['rating'].value_counts()
```

```python
#df['rating']=df['rating'].astype('float32')  : -->could not convert string to float: '|'

df['rating']=df['rating'].str.replace('|',df['rating'].mode()[0]).astype('float32')
```

```python
# Filling null values in rating_count column with mode

df['rating_count']=df['rating_count'].fillna(df['rating_count'].mode()[0])

df['rating_count']=df['rating_count'].str.replace(',','').astype('int32')
```

```python
df.isnull().sum()
```

```python
print(df['category'][0])
df['category'][0]
```

```python
# Here Im going to separate different Values stored in Category Column 

category=df['category'].str.split('|',expand=True)
category.head()
```

```python
category=category.rename(columns={0:'Category_1',1:'Category_2'})

# Adding 2 new categories to df from category

df['category_1']=category['Category_1']
df['category_2']=category['Category_2']
```

```python
df['category_1'].value_counts()
```

```python
df['category_1']=df['category_1'].str.replace('&',' & ')
```

```python
df['category_2'].value_counts()
```

```python
df['category_2']=df['category_2'].str.replace('&',' & ').str.replace(',',', ')
```

```python
# Creating New DataFrame named 'df1'

df1=df[['product_id','product_name','discounted_price','actual_price','discount_percentage',
                       'rating','rating_count','category_1','category_2']]
df1.head()
```

```python
# There are blank spaces present in product_id column

df1['product_id'].str.strip() # -------> for removing blank spaces
```

```python
#Creating Difference of Price Column between Actual Price and Discounted Price

df1['Given_Discount']=df1['actual_price']-df1['discounted_price']
```

```python
# Creating New Categorical columns on the basis of Rating:

ratings=[]
for i in df1.rating:
    if i<2.0: ratings.append('Awful')
    elif i<3.0: ratings.append('Below Average')
    elif i<4.0: ratings.append('Average')
    elif i<5.0: ratings.append('Above Average')
    elif i==5.0: ratings.append('Excellent')
```

```python
df1['Performance']=ratings
df1['Performance']=df1['Performance'].astype('category')
```

# Prediction and Machine Learning

So I gonna use Different Regression techniques with Actual Price and Discounted Price.

Where Actual Price is Independent Variable and Discounted Price is Dependent Variable

```python
#Extracting Independent and Dependent Variables

X = df1.iloc[:, 3].values.reshape(-1, 1)
y = df1.iloc[:, 2].values.reshape(-1, 1)
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
#Calculating the Coefficients

reg.coef_
```

```python
#Calculating the Intercept

reg.intercept_
```

```python
#Calculating the R Squared Value

from sklearn.metrics import r2_score
y_pred = reg.predict(X_test)
print('R2 Score: ', r2_score(y_test, y_pred))
```

```python
#Scatter Plot with Regression Line

plt.scatter(X_train, y_train)
plt.plot(X_test, y_pred, color='red')
plt.xlabel('Actual Price', fontweight='bold')
plt.ylabel('Discounted Price', fontweight='bold')
plt.title('Plot with Regression Line', fontweight='bold')
plt.show()
```

```python
#Cross Validation Result

from sklearn.model_selection import cross_val_score, KFold

kf = KFold(n_splits=10, shuffle=True, random_state=21)

cv_results = cross_val_score(reg, X, y, cv=kf)

print('Cross Validation Results Mean: ', cv_results.mean())
```

```python
#Lasso Regression

from sklearn.linear_model import Lasso

Xl = df1[['actual_price', 'rating', 'rating_count']]
yl = df1['discounted_price']

Xl_train, Xl_test, yl_train, yl_test = train_test_split(Xl, yl, random_state = 21, test_size = 0.3)

lasso = Lasso(alpha = 0.1)
lasso.fit(Xl_train, yl_train)
ridge_predict = lasso.predict(Xl_test)
print('Lasso score: ',lasso.score(Xl_test, yl_test))
print('Lasso coefficients: ',lasso.coef_)
```

## Sentiment Analysis

```python
#Subsetting Reviews
#importing packages for Sentiment Analysis

import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

sia = SentimentIntensityAnalyzer()
sentiment = df[['product_id','review_title','review_content']]
```

```python
#Running the polarity Score on the entire dataset
res = {}

for i, row in tqdm(sentiment.iterrows(), total=len(sentiment)):
    review_desc =row['review_content']
    myid = row['product_id']
    res[myid] = sia.polarity_scores(review_desc)
```

```python
pol_score = pd.DataFrame(res).T
pol_score = pol_score.reset_index().rename(columns={'index' : 'product_id'})

pol_score
```

```python
#Merging the Dataframe with Polarity Score

df4 = df1.merge(pol_score, how='left')
```

```python
#Statistical Description on the Sentiment Score

sentiment_score = df4[['pos', 'neu', 'neg', 'compound']]
sentiment_score.describe()[1:7]
```
